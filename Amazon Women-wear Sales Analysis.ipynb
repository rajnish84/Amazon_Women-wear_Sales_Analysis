```python
# 3 Is there a correlation between 'Qty' and 'Amount'?
#correlation value between Amount and Quantity
sns.displot(df['Amount'],bins=30)
plt.show()
print(df['Amount'].corr(df['Qty']))

#scatter plot
plt.scatter(df['Amount'], df['Qty'])
plt.title('The correlation between Amount and quantity', pad=20)
plt.xlabel('Amount')
plt.ylabel('Quantity')
plt.show()

print('low positive correlation')
```

```python
# 4 What is the most common 'ship-city' in the dataset?
s_city=df['ship-city'].value_counts()[:10]
print('top 10 city with orders:',s_city)
palette_color = sns.color_palette('bright') 
  
# plotting data on chart 
plt.pie(s_city,labels=s_city.index,autopct='%.0f%%',colors=palette_color)
plt.show()
sns.barplot(x=s_city.index, y=s_city.values, palette='viridis')
plt.xticks(rotation=90)
```

```python
# 5 How many unique categories are there in the 'Category' and 'Size' column?

df1=df[['Category', 'Size','Qty']]
pivot=pd.pivot_table(df1, index=['Size'], aggfunc='count')
print(pivot)


pivot1=pd.pivot_table(df1, index=['Category'], aggfunc='count')
pivot1
```

```python
# 6 What is the average 'Amount' for each 'Category
average_amount_category = df.groupby('Category')['Amount'].mean().sort_values(ascending=False)
print(average_amount_category)
plt.figure(figsize=(18,10))
sns.barplot(x=average_amount_category.index, y=average_amount_category.values, palette='viridis')
print('Set has the highest average order value, its almost triple that of Dupatta, inventory should be arranged keeping this in mind.')
```

```python
#7 Top 10 states with highest order value
lst=df['ship-state'].value_counts().sort_values(ascending=False).index[:10]
df2=df.loc[df['ship-state'].isin(lst)]
sns.barplot(x = df2['ship-state'],y = df2['Amount']) #### order function
plt.xticks(rotation=90)
plt.show()

print('Biggest buyers are from the northen region of india i.e U.P,Delhi and Gujarat,Marketing strategy should be adjusted accordingly')

```

```python
#8 What is the distribution of orders over time?

df['Date'] = pd.to_datetime(df['Date'])
orders_per_day = df.resample('D', on='Date').size()

# Plotting the distribution of orders over time
plt.figure(figsize=(12, 6))
plt.plot(orders_per_day, marker='o', linestyle='-', color='b')
plt.title('Distribution of Orders Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Orders')
plt.grid(True)
plt.show()

print('Ther is a decrease in number of orders,find reasons, maybe we need to alter marketing strategy ')
```

```python
#10 average order amount by fulfiment
average_full_amount_category = df.groupby('Fulfilment')['Amount'].mean().sort_values(ascending=False)
print(average_full_amount_category)

print('No major difference found between average order amount')
```
